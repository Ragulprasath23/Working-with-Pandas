### Filtering Data

```python
# Identifying the desired condition and passing it through the dataframe

'''This results in a boolean result, but when again passed into the
dataframe it gives the actual dataframe with the desired result'''

df_pbl['Gender']=='Man'
```

```python
# Filtering the dataframe with 'Gender' column with value as 'Man'

df_pbl[df_pbl['Gender']=='Man']
```

```python
# Filtering data with .loc 
# Assingning the filter to a variable and then passing it to the .loc

filt=df_pbl['Gender']=='Man'
df_pbl.loc[filt,['Hobbyist','Country','Gender']]
```

### Use of (&) and (|) operator

```python
# Using & and | operator to specify conditions

filt1 = (df_pbl['Gender']=='Woman') & (df_pbl['Country']=='United States') 
df_pbl.loc[filt1,['Country','Student','SocialMedia']]
```

```python
filt2 = (df_pbl['Gender']=='Woman') | (df_pbl['Country']=='United States') 
df_pbl.loc[filt2,['Country','Student','SocialMedia']]
```

```python
# Using a (-) to filter out the condition, all results except the filter 

df_pbl.loc[-filt1,['Country','Student','SocialMedia','Gender']]
```

### Groupby 

```python
# Grouping the data based on the country

country_grp  = df_pbl.groupby(['Country'])
```

```python
country_grp
```

```python
df_pbl["Country"].unique()
```

```python
# Getting the grouped values from the dataframe with the get_group()

country_grp.get_group('India')
```

```python
# Filtering the count of social media users in India - (without Groupby)

con_fil = df_pbl['Country'] == 'India'
df_pbl.loc[con_fil,'SocialMedia'].value_counts()
```

```python
# Finding the SocialMedia users per each platform for every country

country_grp['SocialMedia'].value_counts().head(50)
```

```python
# Finding individual data by using groupby column index 
# Filtering the count of social media users in India - (without Groupby)
# Groupby column data becomes the index

country_grp['SocialMedia'].value_counts().loc['India']
```

```python
# Finding median and mean for canada and United States on the grouped data

country_grp['SalaryUSD'].agg(['median','mean']).loc[['Canada','United States']]
```

### Apply function

```python
# Count of people who have worked with python in each country
# Using a lambda function to define the condition 

country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```
