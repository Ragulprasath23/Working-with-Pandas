### Converting the Date (string format) to a Datetime format

pandas.to_datetime() - converts string datatype to datetime datatype

```python
# Converting the Date to Datetime datatype

df['Date'] = pd.to_datetime(df['Date'])
```

Above Code throws an error, as the formatting is not defined and pandas try to figure it out by itself

```python
# Mentioning the format of the datetime while converting
# Formatting values can be identified and referred from pandas date-time documentation

df['Date'] = pd.to_datetime(df['Date'],format='%Y-%m-%d %I-%p')
```

```python
# Verifying the datatype of Date

df.dtypes
```

```python
# Identifying the day name using day_name()

df.loc[10,'Date'].day_name()
```

```python
# identifying the day nmae for the entire date column

df['Date'].dt.day_name()
```

```python
# Identifying the start and end date using min and max

print(df['Date'].min())
print(df['Date'].max())
```

Difference between dates

```python
df['Date'].max() - df['Date'].min()
```

### Selecting data using the date

```python
# Taking selective data from dataframe using a date
# Type-1

filt = (df['Date'] >= '2019') & (df['Date'] < '2020')
df.loc[filt]
```

```python
# Type-2
# Utilizing the to_datetime function

filt2 = df['Date'] > pd.to_datetime('2020-01-01')
df.loc[filt2]
```

As the date and time are unique for each value we can use it as index

```python
# Setting the date to index

df.set_index('Date',inplace=True)
```

```python
df
```

```python
# Selecting all data within the year 2019

df.loc['2019']
```

```python
# Selecting more precise data with more accurate date

df.loc['2020-01':'2020-03']
```

### Aggregate functions with a range of data


```python
# Identifying the mean of Closing value witin the limited time

df.loc['2020-01':'2020-03','Close'].mean()
```

```python
# Getting the max closing value in a  day

df.loc['2020-01-01','Close'].max()
```

```python
# Getting each days max value from high
# resample() - function used on datetime to selcet intervals (day,week)

df['High'].resample('D').max()
```

### Plotting the data 

```python
# Plotting the Max values of the High column

highs = df['High'].resample('D').max()
highs.plot(title='Max of High',ylim=[0,1500])
```
